FFT in Python

```python
import numpy as np
t = np.arange(100) #np array range 0 to 100
s = np.sin(0.15*2*np.pi*t) # the sine wave
```

```python
S = np.fft.fft(s) #fft function
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t,s,'.-') # plt.plot(x_axis, y_axis, marker, ms=20(marker size), mec = 'r'(color))
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("Sine Wave")
```

```python
S_mag = np.abs(S) # a function for magnitude of a complex number
S_phase = np.angle(S) # function for phase
plt.figure(0)
plt.plot(t,S_mag,'.-') 
plt.xlabel("FFT index")
plt.ylabel("FFT of signal (Magnitude)")

# plt.show()
plt.figure(1)
plt.plot(t,S_phase,'.-')
plt.xlabel("FFT index")
plt.ylabel("FFT of signal (Phase)")
```
